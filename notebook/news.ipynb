```python
import requests
from transformers import pipeline
from newsapi import NewsApiClient
```

```python
# Obtener la clave de la API de NewsAPI desde OpenAI Secret Manager
api_key = "ec1deee45eb64a0a93e6313807943f9e"
```

```python
newsapi = NewsApiClient(api_key='ec1deee45eb64a0a93e6313807943f9e')
```

```python
# Definir los parámetros de búsqueda
query = "bitcoin"
language = "en"
```

```python
# Hacer la solicitud a la API de NewsAPI
url = f"https://newsapi.org/v2/everything?q={query}&language={language}&apiKey={api_key}"
response = requests.get(url)
```

```python
# articles = response.json()["articles"]
articles = newsapi.get_everything(q='bitcoin',                                    
                      from_param='2023-04-01',
                      to='2023-04-11',
                      language='en',
                      sort_by='relevancy')["articles"]
sentiment_pipeline = pipeline('sentiment-analysis', framework='tf', model='nlptown/bert-base-multilingual-uncased-sentiment')
sentiments = []
for article in articles:
    title = article["title"]
    description = article["description"]
    url = article["url"]
    published_at = article["publishedAt"]
    
    # Analizar el contenido del artículo
    content = article["content"]
    if content:
        # Preprocesar el contenido
        content = content.replace("\n", " ")
        
        # Analizar el sentimiento del contenido
        sentiment = sentiment_pipeline(content)[0]["label"]
        sentiments.append(sentiment)
        print(f"Title: {title}")
        print(f"Description: {description}")
        print(f"URL: {url}")
        print(f"Published at: {published_at}")
        print(f"Sentiment: {sentiment}\n")
```
